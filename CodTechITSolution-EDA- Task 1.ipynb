```python
#importing country code data
d_country = pd.read_excel("D:/Country-Code.xlsx")
d_country.head()
```

```python
#combining the data frame as in the first dataset there is country code column and in this also we can find the country code
#so ,we can merge dataframe using merge 
pd.merge(data,d_country,on='Country Code',how='left')
```

```python
#saving the new data
new_data=pd.merge(data,d_country,on='Country Code',how='left')
new_data.head(2)
```

```python
county_value= new_data.Country.value_counts().values
county_value
```

```python
#ploting pie chart,to see which country has highest transactions 
plt.pie(county_value,labels=country_name)
```

```python
#above plot is very bad why because ,in other counties zomato is very less ,now we can see only the top 5 transaction 
#so
plt.pie(county_value[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
new_data.groupby(['Aggregate rating', 'Rating color','Rating text'])
#now the data frame is formed
```

```python
ratings=new_data.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"}) #rename :changing 0 to new name 
ratings
```

## observations 
1) when rating is between 4.5 - 4.9 ---> excellent 
2) when rating is between 4 - 3.4 --->very good
3) when rating is between 3.5 to 3.9 ----> good
4) when rating is between 3 to 3.4  and 2.5 to 2.9 ---->average 
5) when rating is between2 to 3.4 ----> poor '''

```python
#visualization 
#barplot 
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
 #Remove duplicates
new_data.drop_duplicates(inplace=True)
#Finding top 10 cuisines 
cuisine_counts =new_data['Cuisines'].value_counts()
cuisine_counts 
```

```python
# Drop rows with mi
# Display the top 10 cuisines
print(cuisine_counts.head(10).reset_index())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the top 10 cuisines
top_cuisines = cuisine_counts.head(10)

plt.figure(figsize=(20, 10))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis')
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```
